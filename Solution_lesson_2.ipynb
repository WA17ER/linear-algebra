```python
def matrix_mult(mat_1, mat_2):
    result = [[sum(a*b for a,b in zip(mat_1_row, mat_2_col)) for mat_2_col in zip(*mat_2)] for mat_1_row in mat_1]
    return result


a = [[1,-2], [3,0]]
b = [[4,-1],[0,5]]
c= matrix_mult(a,b)
print(c)
```
